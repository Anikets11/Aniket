```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# matplotlib :- It is used to plot different chart and graph on the basis of data available
```

```python
X = np.arange(1,11)
X
```

```python
Y = np.random.randint(1,21,10)
Y
```

```python
plt.plot(X,Y) # Used to plot line plot
```

```python
 plt.plot(X,Y, marker ='o', markersize = 10) # marker is used to highlight the points on chart/graph
    # markersize(ms) used to give size of dots
```

```python
plt.plot(X,Y, marker ='o', markersize = 10, linestyle = ':', linewidth=5) #linestyle(ls) is used to make dotted line of plot
# linewidth(lw) Used to increase width of the line
```

```python
plt.plot(X,Y, marker ='o', markersize = 10, linestyle = ':', linewidth = 5, color = 'red', markerfacecolor = 'blue') 
#color(c):- Used to specify color of plot, markerfacecolor:- Used to change change the color of dots/points
```

```python
plt.plot(X,Y, marker ='o', markersize = 10, linestyle = ':', linewidth = 5, color = 'red', markerfacecolor = 'blue')
plt.xlabel('Roll no.')
plt.ylabel('Marks')
plt.title('Student Info') # Used to provide info of plot
```

```python
plt.scatter(X,Y, s = 25) # Used to plot scatter plot, s:- used to specify the size of point/dot
```

```python
plt.scatter(X,Y, s = 50, marker = 'x', c = 'red') # x - for cross, ^ -  for triangle
```

```python
# Draw one complete (1) sine wave
X = np.linspace(0,6.28,49)
Y = np.sin(X)
```

```python
X,Y
```

```python
plt.scatter(X,Y)
```

```python
Z = np.cos(X)
```

```python
plt.scatter(X,Z)
```

```python
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.plot([0,6.28],[0,0])
```
